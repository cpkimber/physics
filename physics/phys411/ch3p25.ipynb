```python
t    = linspace(1e-15,2,1000)

conk = (1/t)**2 * (e**(1/t))/((e**(1/t) - 1)**2)
```

```python
plt.figure(figsize = (14,10))
plt.plot(t,conk, label =r'$\frac{C}{Nk} = \frac{e^{1/t}}{t^2 (e^{1/t} - 1)^2}$')
plt.xlabel("kT/$\epsilon$")
plt.ylabel("C/Nk")
plt.title("Chapter 3 Problem 25e \nCameron Kimber")
plt.legend()
plt.savefig("325.png")
```
